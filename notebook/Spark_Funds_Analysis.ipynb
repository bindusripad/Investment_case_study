# Checkpoint 4: Sector Analysis 1

```python
# read from mapping.csv
mapping = pd.read_csv('../data/mapping.csv',encoding="ISO-8859-1")
# drop rows where category_list is na
mapping.dropna(axis=0, how='any', thresh=None, subset=['category_list'], inplace=True)
```

```python
# clean up the primary sector column
import re
#replace 0 in the beginning of a category with 'Na'
mapping['category_list'] = mapping['category_list'].apply(lambda x:re.sub('^0','Na',x))
#replace all other 0s in category with 'na'
mapping['category_list'] = mapping['category_list'].apply(lambda x: re.sub('0','na',x)) 
```

```python
# for each row in the mapping dataframe, check for the column that has value of 1 and capture the column name
new_map = mapping.loc[:, mapping.columns != 'category_list']
new_map
mapping['Main_sector'] = new_map.idxmax(axis=1)
mapping.shape

```

```python
mapping = mapping[['category_list','Main_sector']]

#mapping = mapping.to_dict('records')
map_dict = dict(zip(mapping.category_list, mapping.Main_sector))
map_dict
```

```python
# extract primary sector from each category in category_list in master_frame
master_frame['Primary_sector'] = master_frame['category_list'].str.split('|').str[0]
master_frame.head()
```

```python
# add Main_sector column to the master_frame - lookup the mapping dictionary for Primary_sector and get the Main_sector
master_frame['Main_sector'] = master_frame['Primary_sector'].map(map_dict)
master_frame.head()
```
