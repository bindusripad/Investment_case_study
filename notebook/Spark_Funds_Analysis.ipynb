# Checkpoint 4: Sector Analysis 1

```python
# read from mapping.csv
mapping = pd.read_csv('../data/mapping.csv',encoding="ISO-8859-1")
# drop rows where category_list is na
mapping.dropna(axis=0, how='any', thresh=None, subset=['category_list'], inplace=True)
```

```python
# clean up the primary sector column
import re
#replace 0 in the beginning of a category with 'Na'
mapping['category_list'] = mapping['category_list'].apply(lambda x:re.sub('^0','Na',x))
#replace all other 0s in category with 'na'
mapping['category_list'] = mapping['category_list'].apply(lambda x: re.sub('0','na',x)) 
```

```python
# for each row in the mapping dataframe, check for the column that has value of 1 and capture the column name
new_map = mapping.loc[:, mapping.columns != 'category_list']
new_map
mapping['Main_sector'] = new_map.idxmax(axis=1)
mapping.shape

```

```python
mapping = mapping[['category_list','Main_sector']]

#mapping = mapping.to_dict('records')
map_dict = dict(zip(mapping.category_list, mapping.Main_sector))
map_dict
```

```python
# extract primary sector from each category in category_list in master_frame
master_frame['Primary_sector'] = master_frame['category_list'].str.split('|').str[0]
master_frame.head()
```

```python
# add Main_sector column to the master_frame - lookup the mapping dictionary for Primary_sector and get the Main_sector
master_frame['Main_sector'] = master_frame['Primary_sector'].map(map_dict)
master_frame.head()
```

# Checkpoint 5: Sector Analysis 2


```python
# D1 : dataframe for USA 
master_frame['raised_amount_usd'].fillna(0, inplace=True)
D1 = master_frame.loc[master_frame['country_code'] == "USA"]

# Investments range is between 5M to 15M
D1 = D1[D1['raised_amount_usd'].between(5000000, 15000000, inclusive=True)]

# find the number of investments in each main_sector
D1['Inv_Count'] = D1.Main_sector.groupby(D1.Main_sector).transform('count')

# find the amount of investments in each main_sector
D1['Inv_Sum'] = D1.groupby('Main_sector')['raised_amount_usd'].transform('sum')

D1.head()


#myList = ['Entertainment', 'Others', 'Manufacturing', 'Automotive & Sports', 'News, Search and Messaging', 'Social, Finance, Analytics, Advertising', 'Cleantech / Semiconductors', 'Health' ]

#D1[~D1['Main_sector'].isin(myList)].shape

```

```python
# Total number of investments for USA

tot_investments = D1.groupby(['Main_sector', 'Inv_Count'], as_index=False).count()
print(tot_investments['Inv_Count'].sum())
```

```python
# Total amount of investments

tot_sum = D1.groupby(['Main_sector','Inv_Sum'], as_index=False).count()

print(tot_sum['Inv_Sum'].sum())
```

```python
# find the main sector with max number of investments
tot_investments.groupby('Main_sector')['Inv_Count'].sum().sort_values(ascending=False)
```
